## Investigation Overview

> my main goal is to see the effect of covid-19 (quarantine) on the system by comparing after and before data for the same month (2019 - 2020)

## Dataset Overview

> The dataset contain data about ride renting system each row represent a ride and columns give details about each ride (time (year , month and day), duration , location)

```python
# import all packages and set plots to be embedded inline
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sb

%matplotlib inline

# suppress warnings from final output
import warnings
warnings.simplefilter("ignore")
```

```python
# load in the dataset into a pandas dataframe
df = pd.read_csv("data.csv")
df.head()
```

## (Change in ride numbers per month)

>from the visual below it it obvious that the system was growning (jan and feb) 2019 vs 2020 but the situation changed in the panadmic months

```python
sb.countplot(x="month", hue="year", data=df).set(title ="Rides count per month for 2019 and 2020");
```

## (changes in riding habbits)

> from the visual it is obvios that the ride numbers per weekends (sat and sun) are higher in 2019 vs 2020 in 2019(before the pandamic ) all days vs 2020 were higher which may indicate that user used the system as workout or for entertainment 

```python
sb.countplot(x="day", hue="year", data=df).set(title ="Rides count per day for day for 2019 and 2020");
```

## (User Type)
> the fig below shows that mostly people in 2019 relied on subscribtion but in 2020 as nothing was clear in the global situation and many worked from home customer methoed arised



```python
sb.countplot(x="user_type", hue="year", data=df).set(title ="Rides count per user type for 2019 and 2020");
```

## (User Type vs month)
> shows the change in user type when covid-19 started (march 2020) vs the year before 


```python
fig, axes = plt.subplots(nrows=1, ncols=2, figsize=(15,5))
fig.suptitle('Rides count per month')
plot2020=sb.countplot(x="month", hue="user_type", data=df[df.year == 2020],ax=axes[0]);
plot2020.set_title('2020');
plot2019 = sb.countplot(x="month", hue="user_type", data=df[df.year == 2019],ax=axes[1]);
plot2019.set_title('2019');
```

## (Ride duration vs month)
> shows that for 2020 the duration time increased during the pandamic which may indicates that people used it to exercise 
however in 2019 the duration was almost the same for all mmonths

```python
fig, axes = plt.subplots(1, 2, figsize=(20, 10))
fig.suptitle('Rides duration per month')
plot2020 =sb.boxplot(ax=axes[0], data=df[df.year == 2020], x='duration_min', y='month');
plot2020.set_title('2020');
plot2019 = sb.boxplot(ax=axes[1], data=df[df.year == 2019], x='duration_min', y='month');
plot2019.set_title('2019');
```
